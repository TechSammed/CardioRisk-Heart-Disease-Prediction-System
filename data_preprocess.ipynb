```python
import pandas as pd

df = pd.read_csv("heart.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["id", "dataset", "ca", "thal", "slope"], inplace=True)
```

```python
df["target"]=(df["num"]>0).astype(int)
df.drop(columns=["num"],inplace=True)
```

```python
print(df["sex"].value_counts())
print(df["fbs"].value_counts())
```

```python
df["sex"] = df["sex"].map({
    "Male": 1,
    "Female": 0
})
```

```python
df["fbs"] = df["fbs"].map({
    True: 1,
    False: 0
})
```

```python
df["cp"] = df["cp"].astype("category").cat.codes
df["restecg"] = df["restecg"].astype("category").cat.codes

```

```python
df=df.apply(pd.to_numeric,errors="coerce")
df.fillna(df.median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv("heart_clean.csv",index=False)
```
